```python
import pandas as pd
import ast
import os
```

```python
reviews_df = pd.read_csv("../../data/top_100_game_reviews.csv")
game_details_df = pd.read_csv("../../data/top_100_game_details.csv")
```

```python
reviews_df.head()
```

```python
game_details_df[game_details_df['appid'] == 730].head()
```

```python
review_and_game_df = reviews_df.merge(game_details_df, on='appid')[['review', 'appid', 'name', 'categories', 'content_descriptors']]
```

```python
review_and_game_df.head()
```

```python
review_and_game_df['categories_str'] = review_and_game_df['categories'].apply(lambda x: ' '.join([y['description'] for y in ast.literal_eval(x)]))
review_and_game_df['content_descriptors_str'] = review_and_game_df['content_descriptors'].apply(lambda x: ast.literal_eval(x)['notes'])
review_and_game_df = review_and_game_df.dropna(subset=['review'])
```

```python
review_and_game_df.head()
```

```python
review_and_game_df['text'] = review_and_game_df.apply(lambda x:
                f"Game Name: {x['name']}\n"
                f"App ID: {x['appid']}\n"
                f"Categories: {x['categories_str']}\n"
                f"Content Descriptors: {x['content_descriptors_str']}\n"
                f"Review: {x['review']}",
    axis=1
)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import Dataset

dataset = Dataset.from_pandas(review_and_game_df[["text"]])

model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(model_name)
model = model.to('cuda')

def tokenize_function_with_labels(examples):
    tokenized = tokenizer(
        examples["text"],
        padding="max_length",
        truncation=True,
        max_length=512
    )
    # Use input_ids as labels for causal language modeling
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

tokenized_dataset = dataset.map(tokenize_function_with_labels, batched=True)
```

```python
train_test_split_dataset = tokenized_dataset.train_test_split(test_size=0.2)
train_dataset = train_test_split_dataset["train"]
eval_dataset = train_test_split_dataset["test"]
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=2,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=2,
    logging_dir="./logs",
    fp16=True
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

trainer.can_return_loss = True
trainer.train()
```

```python
model.save_pretrained("./fine_tuned_model")
tokenizer.save_pretrained("./fine_tuned_model")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_path = "./fine_tuned_model"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path)
```

```python
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device=0
)
```

```python
prompt = "Call of duty is a similar"
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(model.device)

outputs = model.generate(
    input_ids,
    max_length=100,
    num_return_sequences=1,
    do_sample=True,
    top_p=0.95,
    temperature=0.9
)

generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```
